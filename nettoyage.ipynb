### diffÃ©rence entre les deux databases

   afin de fusionner les deux databases il faut dans un premier temps noter leur diffÃ©rences


```python
columns_2015 = list(df_2015.columns)
columns_2016 = list(df_2016.columns)
columns_idem = []
diff_col_2015 = []
diff_col_2016 = []

for col in columns_2015:
    if col in columns_2016:
        columns_idem.append(col)
    else:
        diff_col_2015.append(col)
for col in columns_2016:
    if col in columns_2015:
        columns_idem.append(col)
    else:
        diff_col_2016.append(col)

print('columns in df_2015 and not in df_2016 ',len(diff_col_2015),': \n', diff_col_2015, '\n \n \n',
      'colums in df_2016 and not in df_2015 ',len(diff_col_2016),': \n',diff_col_2016)
```

### DÃ©but du nettoyage afin de concatener les deux 2 dataframes

    - changer nom des colonnes avec les mÃªmes variables
    - rÃ©arranger les colonnes de location qui sont diffÃ©rentes sur les deux dataframes
    
    Une fois les deux dataframes concatÃ©nÃ©s:
    
    - Suppression des doublons
    - Suppression des colonnes inutiles pour la prÃ©diction


```python
### rÃ©arrangement des colonnes de location

df_2015['Location'].iloc[0]


```

```python
df_2015_inter = df_2015.copy()
df_2015_inter['Location'] = [ast.literal_eval(str(item)) for index, item in df_2015_inter.Location.iteritems()]
df_2015_inter = pd.concat([df_2015_inter.drop(['Location'], axis=1), df_2015_inter['Location'].apply(pd.Series)], axis=1)
df_2015_inter['human_address'] = [ast.literal_eval(str(item)) for index, item in df_2015_inter.human_address.iteritems()]
df_2015_inter = pd.concat([df_2015_inter.drop(['human_address'], axis=1), df_2015_inter['human_address'].apply(pd.Series)], axis=1)
df_2015_inter.head()


```

```python
df_2015_inter = df_2015_inter.rename(columns={"latitude":"Latitude", "longitude":"Longitude",
                                      "address":"Address", "city":"City", 
                                      "state":"State", "zip":"ZipCode"})

df_2015_inter.head()
```

```python
### Suppression et renaming des colonnes du dataframe de 2015

df_2015_inter = df_2015_inter.rename(columns={'GHGEmissions(MetricTonsCO2e)':'TotalGHGEmissions',
                                     'GHGEmissionsIntensity(kgCO2e/ft2)':'GHGEmissionsIntensity',
                                     'Comment':'Comments'})


df_2015_inter.drop(['OtherFuelUse(kBtu)','2010 Census Tracts',
                      'Seattle Police Department Micro Community Policing Plan Areas',
                      'City Council Districts','SPD Beats', 'Zip Codes'], axis=1, inplace=True)

print(df_2015_inter.info(verbose = True))
```

```python
### ConcatÃ©nation des deux dataframe

df_final = pd.concat([df_2015_inter,df_2016])
df_final.head()
print(df_final.info(verbose=True))
```

```python
### suppression des lignes concernant les habitations

df_final['BuildingType'].unique()

#df_final.dropna(df_final[['buildingType '] == 'Multifamily MR (5-9)'])
df_final.drop(df_final.loc[df_final['BuildingType']=='Multifamily MR (5-9)'].index, inplace=True)
df_final.drop(df_final.loc[df_final['BuildingType']=='Multifamily LR (1-4)'].index, inplace=True)
df_final.drop(df_final.loc[df_final['BuildingType']=='Multifamily HR (10+)'].index, inplace=True)
df_final.info(verbose = True)
```

```python
#Suppression des doublons

df_final.drop_duplicates(keep = 'first', inplace=True)
df_final.info(verbose = True)
```

```python
### Suppression des colonnes inutiles


'''
- 24 SiteEUIWN(kBtu/sf) (redondant)
- 26 SourceEUIWN(kBtu/sf) (redondant)
- 28 SiteEnergyUseWN(kBtu) (redondant)
- 37 Comments (inutile)
- 38 ComplianceStatus (inutile)
- 39 Outlier (inutile)
- 19 ThirdLargestPropertyUseType (inutile)
- 20 ThirdLargestPropertyUseTypeGFA (inutile)
- 17 SecondLargestPropertyUseType (inutile)
- 18  SecondLargestPropertyUseTypeGFA (inutile)
- 21  YearsENERGYSTARCertified (inutile)
- 31  Electricity(kBtu) (redondant)
- 33  NaturalGas(kBtu) (redondant)
- 36  DefaultData (inutile)
- 4   PropertyName (inutile)
- 43  City (inutile)
- 44  State (inutile)
'''

colonnes = ['SiteEUIWN(kBtu/sf)', 'SourceEUIWN(kBtu/sf)','SiteEnergyUseWN(kBtu)','Comments','ComplianceStatus','Outlier',
            'ThirdLargestPropertyUseType','ThirdLargestPropertyUseTypeGFA','SecondLargestPropertyUseType',
           'SecondLargestPropertyUseTypeGFA','YearsENERGYSTARCertified','Electricity(kBtu)','NaturalGas(kBtu)',
            'DefaultData','PropertyName','City','State']

df_final.drop(columns = colonnes, axis = 1,inplace = True)
df_final.info(verbose = True)
```

```python
df1 = df_final.copy()
```

```python
# Suppression des lignes oÃ¹ 'TotalGHGEmissions' et 'SiteEnergyUse(kBtu)' sont nuls ou nÃ©gatif

df1.drop( df1[ df1['TotalGHGEmissions'] <= 0.0 ].index, inplace=True)
df1.drop( df1[ df1['SiteEnergyUse(kBtu)'] <= 0.0 ].index, inplace=True)
df1.info(verbose = True)
```

```python
### Exportation du dataframe final pour la recherche de features et la prÃ©diction

df1.to_csv('building-energy-benchmarking_final.csv',sep = ',')
```
